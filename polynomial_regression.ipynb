```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn import metrics

from sklearn.metrics import r2_score

from sklearn.preprocessing import PolynomialFeatures

from sklearn.preprocessing import MinMaxScaler

from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv('train.csv')
```

```python
df.isna().sum()
```

```python
X = df.drop(['force'], axis = 1)

Y = df[['force']]
```

```python
x_train , x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.3, random_state=0)
```

```python
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)

polyreg = make_pipeline(PolynomialFeatures(degree = 9), LinearRegression())

pred = polyreg.fit(x_train, y_train).predict(x_test)

pred
```

```python
print(r2_score(y_test, pred))
```

```python
db = pd.read_csv('test.csv')
```

```python
db.isna().sum()
```

```python
X1 = db
```

```python
X1 = scaler.fit_transform(X1)

result = polyreg.fit(x_train, y_train).predict(X1)

db["force"] = result

to_submit = db[["ID","force"]]

to_submit.to_csv("submission4.csv", index = False)
```
